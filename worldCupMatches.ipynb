```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
data = pd.read_csv("WorldCupMatches.csv")
```

```python
data.dropna(inplace=True)
```

```python
df = pd.DataFrame(data)
```

```python
df["Year"] = df["Year"].astype(int)
```

```python
df["Home Team Goals"], df["Away Team Goals"] = df["Home Team Goals"].astype(int) , df["Away Team Goals"].astype(int)
```

```python
df
```

```python
TurkeyMatches = df[(df["Home Team Name"] == "Turkey") | (df["Away Team Name"] == "Turkey")]
```

```python
newDf = pd.DataFrame(TurkeyMatches)
```

```python
newDf
```

## Türkiyenin maç yaptığı şehirler 

```python
plt.figure(figsize=(10,6))
sbn.scatterplot(x=("Year"),y="City",data=newDf)
#plt.xticks(range(df["Year"].min(), df["Year"].max() + 1, 5))
#plt.xticks(sorted(df["Year"].unique()))
plt.show()
```

```python
newDf["Teams"] = newDf["Home Team Name"] + " vs " + newDf["Away Team Name"]
```

```python
newDf
```

```python
plt.figure(figsize=(10,6))
sbn.barplot(x="Teams", y="Home Team Goals",color="red",data = newDf , label="Home" )
sbn.barplot(x="Teams", y="Away Team Goals",color="blue",data = newDf, label= "Away")
plt.legend()
plt.xticks(rotation=45, ha="right")
plt.show()
```
